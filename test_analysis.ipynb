```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
file = 'data-scraper/data/warsaw_apartments_example.csv'
```

```python
df = pd.read_csv(file)
df.tail()
```

```python

# obliczenie średniej ceny dla każdej dzielnicy
avg_price_by_district = (
    df.groupby('district')['price_clean']
      .mean()
      .sort_values()
)

# wykres
plt.figure(figsize=(12, 6))
plt.bar(
    avg_price_by_district.index,
    avg_price_by_district.values
)

plt.xticks(rotation=45, ha='right')
plt.xlabel('Dzielnica')
plt.ylabel('Średnia cena mieszkania')
plt.title('Średnie ceny mieszkań w Warszawie wg dzielnicy')

plt.tight_layout()
plt.show()
```
